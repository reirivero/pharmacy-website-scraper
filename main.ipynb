```python
import requests
from bs4 import BeautifulSoup
import json
import re

# Función para extraer datos del producto desde una URL
def extract_product_data(url):
    response = requests.get(url)
    if response.status_code != 200:
        return None

    # Parsear el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Buscar el script que contiene 'product_traker_data'
    script_tag = soup.find('script', string=re.compile(r'var product_traker_data ='))
    if not script_tag:
        return None

    # Extraer el contenido del script
    script_content = script_tag.string

    # Extraer el JSON del script
    json_data = re.search(r'var product_traker_data = ({.*});', script_content).group(1)
    product_data = json.loads(json_data)

    # Extraer la información requerida
    name = product_data['name']
    is_available = product_data['isAvailable']
    price = product_data['price']
    bioequivalent = product_data['products'][list(product_data['products'].keys())[0]]['params']['bioequivalent']

    # Convertir isAvailable y bioequivalent a booleanos
    is_available = True if is_available else False
    bioequivalent = True if bioequivalent else False

    # Extraer el "Principio Activo" del meta tag
    meta_description = soup.find('meta', property='og:description')['content']
    active_principle_match = re.search(r'Principio Activo: ([^|]+)', meta_description)
    active_principle = active_principle_match.group(1).strip().split('/')[0].strip() if active_principle_match else None

    # Extraer el SKU de la URL
    sku_match = re.search(r'default_sku=(\d+)', url)
    sku = sku_match.group(1) if sku_match else None

    # Extraer el nombre del laboratorio del bloque de código HTML proporcionado
    lab_name_tag = soup.find('div', class_='description-area').find('h4', sring='Laboratorio')
    lab_name = lab_name_tag.find_next_sibling('p').text.strip() if lab_name_tag else None

    return {
        "name": name,
        "is_available": is_available,
        "price": '$'+price,
        "bioequivalent": bioequivalent,
        "active_principle": active_principle,
        "sku": sku,
        "lab_name": lab_name
    }

# Lista de URLs de productos
# urls = [
#     "https://salcobrand.cl/products/drovelis-drospirenona-estetrol-28-comprimidos-recubiertos?srsltid=AfmBOoptRyxI3ukdc-wTLtABfXaZMrp645sdWPpmHU9Am1hwCMvKo5GY",
#     "https://salcobrand.cl/products/estradiol-0-1-topico-semisolido-1429a6a8-f299-4655-837a-354fbfd3e7a7?default_sku=3028594",
#     # Agrega más URLs aquí
# ]

# Extraer datos de cada URL
# Extraer datos de cada URL
for url in urls:
    try:
        product_info = extract_product_data(url)
        if product_info:
            print(f"URL: {url}")
            print(f"Nombre del producto: {product_info['name']}")
            print(f"Disponibilidad en stock: {product_info['is_available']}")
            print(f"Precio: {product_info['price']}")
            print(f"Es bioequivalente: {product_info['bioequivalent']}")
            print(f"Principio Activo: {product_info['active_principle']}")
            print(f"SKU: {product_info['sku']}")
            print(f"Nombre del laboratorio: {product_info['lab_name']}")
            print("-" * 40)
        else:
            print(f"No se pudo extraer la información del producto en la URL: {url}")
    except Exception as e:
        print(f"Error al procesar la URL {url}: {e}")
```

# Farmacia **Farmacia Dr Simi**



```python
from bs4 import BeautifulSoup
import requests

# URL del producto
drsimi_urls = [] 
u1= 'https://www.drsimi.cl/gestel-200-mg-progesterona-micronizada/p'
u2 = 'https://www.drsimi.cl/citrato-de-magnesio-200mg-60cap-ch6112/p' 

drsimi_urls.append(u1)
drsimi_urls.append(u2)

for url in drsimi_urls:
    # Realizar la solicitud HTTP
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontrar el contenedor del precio
    price_container = soup.find('span', class_='vtex-product-price-1-x-currencyContainer')

    # Extraer y formatear el precio
    currency_code = price_container.find('span', class_='vtex-product-price-1-x-currencyCode').text
    currency_integer = price_container.find_all('span', class_='vtex-product-price-1-x-currencyInteger')
    currency_group = price_container.find('span', class_='vtex-product-price-1-x-currencyGroup').text

    # Formatear el precio
    normal_price = f"{currency_code}{currency_integer[0].text}{currency_group}{currency_integer[1].text}"

    print("Precio normal: ",normal_price,type(normal_price))  # Debería imprimir: $15.600

    # Encontrar el contenedor del precio con rebaja
    price_container = soup.find('span', class_='vtex-product-price-1-x-sellingPriceValue')

    # Extraer y formatear el precio
    currency_code = price_container.find('span', class_='vtex-product-price-1-x-currencyCode').text
    currency_integer = price_container.find_all('span', class_='vtex-product-price-1-x-currencyInteger')
    currency_group = price_container.find('span', class_='vtex-product-price-1-x-currencyGroup').text

    # Formatear el precio
    price = f"{currency_code}{currency_integer[0].text}{currency_group}{currency_integer[1].text}"

    print("Precio actual: ",price,type(price))  # Debería imprimir: $11.700

    # Encontrar el SKU
    sku_container = soup.find('span', class_='vtex-product-identifier-0-x-product-identifier__value')
    sku = sku_container.text.strip()
    print("SKU: ", sku)

    # Encontrar nombre del producto
    name_container = soup.find('span', class_='vtex-store-components-3-x-productBrand vtex-store-components-3-x-productBrand--quickview')
    name_product = name_container.text.strip()

    print("Nombre del producto: ", name_product)

    # Encontrar el contenedor del texto de bioequivalencia
    bioequivalente_elem = soup.find('p', class_='farmaciasdeldrsimicl-theme-2-x-bioequivalenteText')
    if bioequivalente_elem:
        bioequivalente_text = bioequivalente_elem.text

        # Verificar si el texto contiene "es bioequivalente"
        bioequivalente = "es bioequivalente" in bioequivalente_text
    else: bioequivalente = False

    print("¿Es bioequivalente? ","Sí" if bioequivalente else "No")  # Debería imprimir: True si el medicamento es bioequivalente, de lo contrario False

    # Extraer el principio activo
    principio_activo_elem = soup.find('td', class_='vtex-store-components-3-x-specificationItemSpecifications--principioActivo')
    if principio_activo_elem:
        principio_activo = principio_activo_elem.text.strip()
    else: principio_activo = 'Not found in the webpage'

    print("Principio Activo:", principio_activo)  # Debería imprimir: PROGESTERONA

# normal_price, price, sku, name_product, bioequivalente
```

# Farmacia **Eco Farmacias**


```python
# URL del producto
ecof_urls = [] 
u1= 'https://www.ecofarmacias.cl/producto/gestel-capsulas-blandas-100-mg-progesterona/'
u2 = 'https://www.ecofarmacias.cl/producto/progendo-200-mg-x-30/' 

u3 = 'https://www.ecofarmacias.cl/producto/hiedroferol-0266-mg-calcifediol/'
u4 = 'https://www.ecofarmacias.cl/producto/acuode-colecalciferol-50-000-ui-polvo-oral-2-sobres/'
ecof_urls.append(u1)
ecof_urls.append(u2)
ecof_urls.append(u3)
ecof_urls.append(u4)

for url in ecof_urls:
    # Realizar la solicitud HTTP
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extraer el precio
    price = soup.find('bdi').text

    # Extraer el valor de stock
    stock = False if soup.find('p', class_=re.compile(r'stock')) else True

    # Extraer el SKU
    sku = soup.find('span', class_='sku').text

    # Extraer los principios activos
    principios_activos_elem = soup.find('li', string=lambda x: x and 'Principios Activos:' in x)
    if principios_activos_elem:
        principios_activos = principios_activos_elem.text.split(': ')[1]
    else: principios_activos = None

    # Extraer el nombre del producto y laboratorios
    product_title = soup.find('h1', class_='product_title entry-title').text
    product_name = product_title.split('(')[0].strip()
    laboratorios = [lab.strip(')') for lab in product_title.split('(')[1:]]

    # Imprimir los resultados
    print("Nombre del producto:", product_name)  # Debería imprimir: Gestel 200 mg x30 Cápsulas Blandas
    print("Laboratorios:", laboratorios)  # Debería imprimir: ['Cenabast', 'Exeltis']
    print("Precio:", price)  # Debería imprimir: $13.860
    print("Stock:", stock)  # Debería imprimir: False
    print("SKU:", sku)  # Debería imprimir: 7804656600132
    print("Principios Activos:", principios_activos)  # Debería imprimir: Progesterona micronizada
```
