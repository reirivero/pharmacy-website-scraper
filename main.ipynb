```python
import requests
from bs4 import BeautifulSoup
import json
import re

# Función para extraer datos del producto desde una URL
def extract_product_data(url):
    response = requests.get(url)
    if response.status_code != 200:
        return None

    # Parsear el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Buscar el script que contiene 'product_traker_data'
    script_tag = soup.find('script', string=re.compile(r'var product_traker_data ='))
    if not script_tag:
        return None

    # Extraer el contenido del script
    script_content = script_tag.string

    # Extraer el JSON del script
    json_data = re.search(r'var product_traker_data = ({.*});', script_content).group(1)
    product_data = json.loads(json_data)

    # Extraer la información requerida
    name = product_data['name']
    is_available = product_data['isAvailable']
    price = product_data['price']
    bioequivalent = product_data['products'][list(product_data['products'].keys())[0]]['params']['bioequivalent']

    # Convertir isAvailable y bioequivalent a booleanos
    is_available = True if is_available else False
    bioequivalent = True if bioequivalent else False

    # Extraer el "Principio Activo" del meta tag
    meta_description = soup.find('meta', property='og:description')['content']
    active_principle_match = re.search(r'Principio Activo: ([^|]+)', meta_description)
    active_principle = active_principle_match.group(1).strip().split('/')[0].strip() if active_principle_match else None

    # Extraer el SKU de la URL
    sku_match = re.search(r'default_sku=(\d+)', url)
    sku = sku_match.group(1) if sku_match else None

    # Extraer el nombre del laboratorio del bloque de código HTML proporcionado
    lab_name_tag = soup.find('div', class_='description-area').find('h4', sring='Laboratorio')
    lab_name = lab_name_tag.find_next_sibling('p').text.strip() if lab_name_tag else None

    return {
        "name": name,
        "is_available": is_available,
        "price": '$'+price,
        "bioequivalent": bioequivalent,
        "active_principle": active_principle,
        "sku": sku,
        "lab_name": lab_name
    }

# Lista de URLs de productos
# urls = [
#     "https://salcobrand.cl/products/drovelis-drospirenona-estetrol-28-comprimidos-recubiertos?srsltid=AfmBOoptRyxI3ukdc-wTLtABfXaZMrp645sdWPpmHU9Am1hwCMvKo5GY",
#     "https://salcobrand.cl/products/estradiol-0-1-topico-semisolido-1429a6a8-f299-4655-837a-354fbfd3e7a7?default_sku=3028594",
#     # Agrega más URLs aquí
# ]

# Extraer datos de cada URL
# Extraer datos de cada URL
for url in urls:
    try:
        product_info = extract_product_data(url)
        if product_info:
            print(f"URL: {url}")
            print(f"Nombre del producto: {product_info['name']}")
            print(f"Disponibilidad en stock: {product_info['is_available']}")
            print(f"Precio: {product_info['price']}")
            print(f"Es bioequivalente: {product_info['bioequivalent']}")
            print(f"Principio Activo: {product_info['active_principle']}")
            print(f"SKU: {product_info['sku']}")
            print(f"Nombre del laboratorio: {product_info['lab_name']}")
            print("-" * 40)
        else:
            print(f"No se pudo extraer la información del producto en la URL: {url}")
    except Exception as e:
        print(f"Error al procesar la URL {url}: {e}")
```

```python
options = Options()
# options.headless = True  # Ejecuta el navegador en modo headless
options.add_argument("--headless")  # Ejecutar en modo headless
options.binary_location = "/usr/bin/google-chrome"  # Ruta al binario de Google Chrome
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=options)

urls = list()
url1 = 'https://farmaciaelquimico.cl/products/hormogel-pum-gel-transdermico-0-5mg?srsltid=AfmBOoowigxqRufUqrN5jS1Lk8WhE9QSlQPbXXQD3h94Lt2oDf7VBzJV'
url2 = 'https://farmaciaelquimico.cl/products/gestel-capsulas-blandas-200mg?srsltid=AfmBOort0qkM4twZDgesx6X1YaxseitLP8Ab3b_3LAwuefWI_X2EJBsV'

urls.append(url1)
urls.append(url2)

products = {
    'date' : [],
    'sku' : [],
    'product_name' : [],
    'price' : [],
    'pharmacy': [],
    'stock': []
}

for url in urls:
    driver.get(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')    
    wait = WebDriverWait(driver, 10)
    price_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.money-subtotal')))
    
    # Avaiibity - stock 

    stock_elem = soup.find(lambda tag: tag.name == 'span' and tag.get('class') == ['productView-info-value'] and 'En stock' in tag.text)
    # stock_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.productView-info-value')))
    elem_product = soup.find('h1',{'class':'productView-title'})
    name_product = elem_product.get_text().strip() if elem_product \
        else "Name product not found"    
    price = price_element.text
    stock = True if stock_elem.text.strip() == 'En stock' else False # True if stock_element.text.strip() == 'En stock' else False
    # print(f'Precio: {price}')
    # print(f'Stock: {stock}')
    # print(f'Nombre: {name_product}')
    sku_elem = soup.find_all('span',class_='productView-info-value')
    sku = (list(sku_elem))[1].text.strip()
    print("SKU: ", sku)

    # Encontrar el contenedor del principio activo
    tab_content = soup.find('div', class_='tab-popup-content')
    principio_activo = tab_content.find('strong').text
    print("Principio activo: ", principio_activo)

    # Encontrar el contenedor del vendedor
    vendor_container = soup.find('div', class_='productView-info-item')
    vendor = vendor_container.find('span', class_='productView-info-value').text.strip()
    
    print("Laboratorio: ", vendor)
    
    # Tiempo actual datetime
    current_time = datetime.now()

    # Actualizando el diccionario de datos
    products['date'].append(current_time)
    products['sku'].append(sku)
    products['product_name'].append(name_product)
    products['price'].append(price)
    products['pharmacy'].append('El Químico')
    products['stock'].append(stock)    

driver.quit()
df = pd.DataFrame(products)
df.head(20)
```

```python
import requests
from bs4 import BeautifulSoup
import json

def extract_product_data(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise requests.exceptions.HTTPError(f"Error en la solicitud: {response.status_code}")
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encontrar el script con el JSON
    script_tag = soup.find('script', id='__NEXT_DATA__', type='application/json')
    json_data = json.loads(script_tag.string)
    
    # Extraer los datos necesarios
    product_data = json_data['props']['pageProps']['initialProduct']
    
    is_bioequivalent = product_data['isBioequivalent']
    active_principle = product_data['activePrinciple']
    laboratory = product_data['laboratory']
    name = product_data['name']
    price = product_data['price']

    # Cambio en la disponibilidad
    # stock = product_data['availability']['stock']
    # print(stock)
    # # Convertir stock a booleano
    # is_available = stock > 0

    # 2024-09-05
        # Verificar la disponibilidad buscando "Recibe" seguido de "mañana"
    is_available = False
    for p_tag in soup.find_all('p', class_='MuiTypography-root MuiTypography-body1 mui-style-m99pms'):
        if "Recibe" in p_tag.text and "mañana" in p_tag.text:
            is_available = True
            break
    
    # Crear un diccionario con los datos extraídos
    data = {
        "is_bioequivalent": is_bioequivalent,
        "active_principle": active_principle,
        "laboratory": laboratory,
        "name": name,
        "price": price,
        "is_available": is_available
    }
    
    return data

# URL de la página
url = 'https://farmaciameki.cl/producto/gestel-cenabast-200-mg-exeltis-110a'

# Extraer los datos del producto
product_data = extract_product_data(url)
print(product_data)
```
