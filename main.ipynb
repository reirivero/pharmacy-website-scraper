```python
options = Options()
# options.headless = True  # Ejecuta el navegador en modo headless
options.add_argument("--headless")  # Ejecutar en modo headless
options.binary_location = "/usr/bin/google-chrome"  # Ruta al binario de Google Chrome
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=options)

urls = list()
url1 = 'https://farmaciaelquimico.cl/products/hormogel-pum-gel-transdermico-0-5mg?srsltid=AfmBOoowigxqRufUqrN5jS1Lk8WhE9QSlQPbXXQD3h94Lt2oDf7VBzJV'
url2 = 'https://farmaciaelquimico.cl/products/gestel-capsulas-blandas-200mg?srsltid=AfmBOort0qkM4twZDgesx6X1YaxseitLP8Ab3b_3LAwuefWI_X2EJBsV'

urls.append(url1)
urls.append(url2)

products = {
    'date' : [],
    'sku' : [],
    'product_name' : [],
    'price' : [],
    'pharmacy': [],
    'stock': []
}

for url in urls:
    driver.get(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')    
    wait = WebDriverWait(driver, 10)
    price_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.money-subtotal')))
    
    # Avaiibity - stock 

    stock_elem = soup.find(lambda tag: tag.name == 'span' and tag.get('class') == ['productView-info-value'] and 'En stock' in tag.text)
    # stock_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.productView-info-value')))
    elem_product = soup.find('h1',{'class':'productView-title'})
    name_product = elem_product.get_text().strip() if elem_product \
        else "Name product not found"    
    price = price_element.text
    stock = True if stock_elem.text.strip() == 'En stock' else False # True if stock_element.text.strip() == 'En stock' else False
    # print(f'Precio: {price}')
    # print(f'Stock: {stock}')
    # print(f'Nombre: {name_product}')
    sku_elem = soup.find_all('span',class_='productView-info-value')
    sku = (list(sku_elem))[1].text.strip()
    print(sku)
    current_time = datetime.now()
    products['date'].append(current_time)
    products['sku'].append(sku)
    products['product_name'].append(name_product)
    products['price'].append(price)
    products['pharmacy'].append('El Químico')
    products['stock'].append(stock)    

driver.quit()
df = pd.DataFrame(products)
df.head(20)
```
